```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
exponential = np.random.exponential([0,1,100])
plt.hist(exponential)
plt.axvline(exponential.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(exponential.mean() + exponential.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(exponential.mean() - exponential.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
beta = np.random.beta([0.5,1],100)
plt.hist(beta)
plt.axvline(beta.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(beta.mean() + beta.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(beta.mean() - beta.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
dirichlet = np.random.dirichlet([0.5,1],100)
plt.hist(dirichlet)
plt.axvline(dirichlet.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(dirichlet.mean() + dirichlet.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(dirichlet.mean() - dirichlet.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
standardExpontential = np.random.standard_exponential(100)
plt.hist(standardExpontential)
plt.axvline(standardExpontential.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(standardExpontential.mean() + standardExpontential.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(standardExpontential.mean() - standardExpontential.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
standardCauchy = np.random.standard_cauchy(100)
plt.hist(standardCauchy)
plt.axvline(standardCauchy.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(standardCauchy.mean() + standardCauchy.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(standardCauchy.mean() - standardCauchy.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
uniform = np.random.uniform(0,1,100)
plt.hist(uniform)
plt.axvline(uniform.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(uniform.mean() + uniform.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(uniform.mean() - uniform.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
rand1 = np.random.normal(5,0.5,100)
rand2 = np.random.normal(10,1,100)
rand3 = rand1 + rand2
plt.hist(rand3)
plt.axvline(rand3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(rand3.mean() + rand3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(rand3.mean() - rand3.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
## Rand3 shows that when you run the graph over and over again, you see some changes in overall data but 
## the mean and standard deviation will relatively stay where they are (mean = ~15, standard deviation = ~1)
```
